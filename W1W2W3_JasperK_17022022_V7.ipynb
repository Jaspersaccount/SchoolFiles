# CLEANING THE DATA

The merged dataframe will now be used to check for any outliers and will be cleaned accordingly.

```python

df_full['gender'].value_counts()
```

```python
df_full['height'].value_counts()
```

```python
df_full['weight'].value_counts()
```

The function below will replace all values below the 45 for the weight segment to a NaN value as these weights will be considered as false data.

```python
def weight_below45(x):
    if (x <= 45) or (x >=300):
        return float('NaN')
    else:
        return x
df_full['weight'] = df_full['weight'].apply(weight_below45)
```

```python
df_full['weight'].value_counts()
```

## PARTICIPANT'S AVERAGES

Below the averages of the participants will be measured

```python
df_full['average steps'] = df_full.iloc[:,9:336].mean(axis=1,skipna=True)
df_full.head()
```

```python
df_full['average steps'].value_counts().sort_index(0)
```

as can be seen in the table above, a new column with average steps was added to to the dataframe.

```python
def averagestepsbelow25k(x):
    if (x >= 25000) or (x <=100):
        return float('NaN')
    else:
        return x
df_full['average steps'] = df_full['average steps'].apply(averagestepsbelow25k)
```

```python
df_full['average steps'].value_counts().sort_index(0)
```

Another if else statement was used to remove outliers in the average step column as these number are asumed to be caused by false data entries and can hinder the future analysis

# LINEAR REGRESSION

This part of the analysis will show the linear regression between variables in the FitBit dataset. this step will generate a formula based on the correlation between two variables. The regression is called linear as the increments to which the trend either rises or falls in remain the same.

```python
pd.get_dummies(data=df_full, columns=['education_1','city','gender','age']) #create dummy variables for all catagorical variables
```

In order to be able to use every single variable in the dataframe the catagorical values where all prepared with dummies using pandas.

```python
sns.lmplot(data=df_full, x='average steps',y='weight')
plt.title('Linear regression between Weight and Average Steps', fontsize=20)
```

```python
#df_full['average steps_2'] = df_full[['average steps']].applymap(np.int64)
df_full['average steps'].isna().sum() #check for NaN values in Average Steps
df_full['weight'].isna().sum() # check for NaN values weight
df_2 = df_full.dropna(subset = ['average steps', 'weight']) #drops the rows with NaN values
df_2.info()
```

```python
v1 = df_2[['weight']]
v2 = df_2['average steps']
lin = LinearRegression()
lin.fit(v1,v2)
v4 = lin.intercept_
v5 = lin.coef_[0]
sns.lmplot(data=df_2,y='average steps',x='weight')
print(f"The Formula for the regression line is Y = {v4:.2f} + {v5:.2f}X")
```

# EVALUATION

in order to evaluate the model and formula the data will first be divided into training and test sets. after which the training set will be used to draw a new model and the test set will be used to evaluate the model.

```python
x_train, x_test, y_train, y_test = train_test_split(v1,v2, test_size = 0.3, random_state=12)
lin_ev = LinearRegression()
lin_ev = lin_ev.fit(x_train, y_train)
```

Builds the different sets and creates a model with the trainings data.

```python
y_testp = lin_ev.predict(x_test)
r2raw = r2_score(y_test, y_testp)
percentR2 = r2raw * 100
print('The R2 score is = '+ str(round(percentR2,2)) +'% which means less than 1 percent of variations is explained by the model')
```

```python
math.sqrt(mean_squared_error(y_test, y_testp))
```

as can be seen by the R2 method the the predictions are and can deviate 2905 steps from the average. what can concluded is that weight is not a viable factor to predict average steps on.

```python
df_2['a steps predict'] = lin_ev.predict(v1)
df_2['residuals'] = df_2['average steps'] - df_2['a steps predict']
```

```python
sns.scatterplot(data = df_2, x='weight', y='residuals')
plt.xlabel = 'Weight in KGs'
plt.ylabel = 'Residuals of Average Steps'
plt.show()
```

what can be seen by plotting the residuals is that they remain very inaccurate throughout the change of size. Also indicating Weight is a bad predictor of Average Steps.
